### Estudio de la variable objetivo 'isFraud'

```python
percent_distribution_target = df['isFraud'].value_counts(normalize=True).mul(100).rename('Percent').reset_index()
count_target = df['isFraud'].value_counts().reset_index()

df_target_distribution = pd.merge(percent_distribution_target, count_target, on=['index'], how='inner').drop('index', axis=1)

print(f'Distribución de la variable objetivo:')
df_target_distribution
```

#### Plot de variable objetivo:

```python
colors = sns.color_palette('pastel')[1:3]

fig, ax = plt.subplots(figsize=(7,7))
plt.pie(df_target_distribution['Percent'], labels = ['No', 'Yes'], colors = colors, autopct='%.2f%%')
plt.title('Percent Distribution of Fraud', fontsize = 20)
plt.show()
```

Vemos claramente que está muy desbalanceado

### Estudiamos las proporciones del tipo de transacciones


```python
plt.figure(figsize=(12,8))
ax=sns.countplot(x = "type", hue="isFraud", data = df)
plt.title('Tipos de transacciones Fraude y No Fraude')
for p in ax.patches:
        ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+50))
```

- Vemos como sólo tienen fraude *TRANSFER* y *CASH OUT*.
- También hay que tener en cuenta que *DEBIT* tiene muchos menos datos que el resto.
- Parece que el número de fraudes de *TRANSFER* y de *CASH OUT*, es casi el mismo. Podemos concluir a priori que modus operandi es hacer una transferencia y luego sacar ese dinero en efectivo.

Vamos a observar si hay fraude en algún dispositivo concreto

```python
print("Transacciones correctas: \n",df[df["isFraud"] == 0].device.value_counts())
print("\n Transacciones fraudulentas: \n",df[df["isFraud"] == 1].device.value_counts())
```

Vemos que las proporciones son bastante parecidas y no podemos sacar conclusiones a priori de estos datos.

Vamos a estudiar la variable step.

```python
df[df["isFraud"] == 0].hist(column="step",color="green",bins=50)
plt.xlabel("Step (unidad = 1 hora)")
plt.ylabel("Número de transacciones")
plt.title("Número de transacciones válidas a lo largo del tiempo")

df[df["isFraud"] == 1].hist(column ="step",color="red",bins=50)
plt.xlabel("Step (unidad = 1 hora)")
plt.ylabel("Número de transacciones")
plt.title("Número de transacciones fraudulentas a lo largo del tiempo")

plt.tight_layout()
plt.show()
```

Vemos como las transacciones válidas ocurren entorno a las 15 horas y a las 40, mientras que las fraudulentas no parecen cambiar mucho a lo largo del tiempo. 
Vamos a estudiar la distribución en los distintos días de la semana.

```python
num_days = 7
num_hours = 24
fraud_days = (df[df["isFraud"] == 1].step//num_hours)% num_days
valid_days = (df[df["isFraud"] == 0].step//num_hours) % num_days

# plotting scatterplot of the days of the week, identifying the fraudulent transactions (red) from the valid transactions (green) 
plt.subplot(1, 2, 1)
fraud_days.hist(bins=num_days,color="red")
plt.xlabel("Días")
plt.ylabel("Número de transacciones")
plt.title("Número de transacciones fraudulentas por día")

plt.subplot(1,2,2)
valid_days.hist(bins=num_days,color="green")
plt.xlabel("Días")
plt.ylabel("Número de transacciones")
plt.title("Número de transacciones válidas por día")

plt.tight_layout()
plt.show()
```

Vemos claramente como las transacciones válidas ocurren la gran mayoría los dos primeros días.

Vamos a estudiar las horas del día donde se producen estas transacciones

```python
valid_hours = df[df["isFraud"] == 0].step % num_hours
fraud_hours = df[df["isFraud"] == 1].step % num_hours

plt.subplot(1, 2, 1)
fraud_hours.hist(bins=num_hours, color="red")
plt.xlabel("Hora")
plt.ylabel("Número de transacciones")
plt.title("Número de transacciones fraudulentas por hora")


plt.subplot(1, 2, 2)
valid_hours.hist(bins=num_hours, color="green")
plt.xlabel("Hora")
plt.ylabel("Número de transacciones")
plt.title("Número de transacciones válidas por hora")

plt.tight_layout()
plt.show()
```

Volvemos a ver una distribución clara en la que de 0 a 8 no se producen casi transacciones válidas. Mientras que las fraudulentas ocurren a casi cualquier hora.

Vamos a estudiar la variable connection_time

Vemos como la columna *Connection_time* está declarada como object porque sus decimales están separados por coma. Hacemos la transformación de la columna a tipo float.

```python
df['connection_time'] = df['connection_time'].apply(lambda x: x.replace(',', '.')).astype(float) # Cambiamos connection_time de object a float
```

```python
df[df["isFraud"] == 0].hist(column="connection_time",color="green",bins=50)
plt.xlabel("Segundos")
plt.ylabel("Número de transacciones")
plt.title("Segundos que tardan en llegar los datos de transacciones válidas")

df[df["isFraud"] == 1].hist(column ="connection_time",color="red",bins=50)
plt.xlabel("Segundos")
plt.ylabel("Número de transacciones")
plt.title("Segundos que tardan en llegar los datos de transacciones fraudulentas")

plt.tight_layout()
plt.show()
```

Aparentemente no hay ningún patrón para poder diferenciar entre las transacciones fraudulentas y válidas.
